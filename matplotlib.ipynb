```python
import matplotlib.pyplot as plt
import numpy as np
anos = [2020, 2021, 2022, 2023, 2024]
vendas = [100, 120, 150, 130, 160]

plt.plot(anos, vendas)
plt.title('Vendas ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Venda (em milhões)')
plt.grid(True)
plt.show()
```

```python
produtos = ['Produto A', 'Produto B', 'Produto C']
quantidades = [50, 80, 65]

plt.bar(produtos, quantidades)
plt.title('Quantidade de Produtos em Estoque')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.show()
```

```python
np.random.seed(42)
x = np.random.rand(50) * 10
y = 2 * x + np.random.randn(50) * 2

plt.scatter(x,y)
plt.title('Relação entre X e Y')
plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.show()
```
