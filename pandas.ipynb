### 2. Estrutura de Dados: Series e DataFrames
- **Séries:** É uma matriz unidimensional com rótulos (índices). Pode ser vista com o uma única coluna de uma tabela.
- **DataFrame:** É uma estrutura bidimensional (linhas e colunas), como uma planilha completa. É essencialmente um conjunto de Series.

```python
# Criando uma Series
notas = pd.Series([8.5, 9.0, 7.5, 10.0], index=['Aluno A', 'Aluno B', 'Aluno C', 'Aluno D'])
print("Exemplo de Series:\n", notas, "\n")
```

```python
# Criando um DataFrame a partir de um dicionário (útil para testes locais)
dados_turma = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Diana'],
    'Idade': [20, 22, 21, 23],
    'Nota_Final': [8.5, 7.0, 9.5, 6.5],
    'Aprovado': [True, True, True, False]
}
df = pd.DataFrame(dados_turma)
print("Exemplo de DataFrame:")
display(df) # O 'display' renderiza uma tabela mais bonita no Jupyter do que o print1
```

### 3. Lendo Dados de Arquivos Externos (CSV)

Na prática real de análise, nós raramente geramos os dados manualmente no código. O fluxo normal é ler os dados diretamente de arquivos, sendo o formato CSV o mais comum.


```python
# O comando abaixo é o padrão para ler arquivos CSV.
# df_csv = pd.read_csv('caminho/para/seu_arquivo.csv')

# Exemplo prático carregando um dataset público diretamente de uma URL para demonstração:
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df_titanic = pd.read_csv(url)
print("Dados carregados com sucesso!")
display(df_titanic)
```

### 5. Seleção e Filtragem de Dados

Como selecionar colunas específicas e filtrar linhas baseadas em condições

```python
# Selecionando apenas uma colunas (retorna uma Series)
idades = df_titanic['Age']
print(idades)

# Selecionando múltiplas colunas (retorna um novo DataFrame)
df_reduzido = df_titanic[['Name', 'Age', 'Survived']]
print(df_reduzido)

# Filtragem (Querying): Selecionando apenas passageiros com mais de 60 anos
passageiros_idosos = df_titanic[df_titanic['Age'] > 60]
print(f"Total de passageiros com mais de 60 anos: {len(passageiros_idosos)}")

# FIltragem com múltiplas condições (Use & para E, | para OU)
# Ex: Mulheres que sobrevivem na 1º classe
filtro = (df_titanic['Sex'] == 'female') & (df_titanic['Survived'] == 1) & (df_titanic['Pclass'] == 1)
mulheres_1a_classe = df_titanic[filtro]

print('Mulheres da 1º classe que sobreviveram:', len(mulheres_1a_classe))
```
