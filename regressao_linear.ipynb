### Importação das Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Carregamento dos dados do CSV

```python
try:
    # Delimitador é vírgula e pula a primeira linha (cabeçalho) com skiprows=1
    dados = np.loadtxt('dataset.csv', delimiter=',', skiprows=1)
except FileNotFoundError:
    print("Erro: Arquivo 'dataset.csv' não encontrado.")
    print("Por favor, crie o arquivo ou execute o script para gerá-lo.")
```

```python
# Apenas visualizando os dados carregados do arquivo CSV.
print(dados)
```

### Separando os dados de X e Y

```python
X = dados[:, 0]
Y = dados[:, 1]

# Visualizando os dados
print("x:",X)
print("y:",Y)

print("\n Número de amostras:", X.shape[0])
```

### Reestruturando os dados

```python
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)

print("X:",X)
print("")
print("Y:",Y)

#Descomente a linha abaixo se quiser ver detalhes os dados

# print("Número de características:", X.shape[1])
# print("Número de saídas:", y.shape[1])
# print("Número total de elementos em X:", X.size)
# print("Número total de elementos em Y:", Y.size)
# print("Tipo de dados em X:", X.dtype)
# print("Tipo dos dados em Y:", Y.dtype)
# print("Tamanho em bytes de cada elemento em X:", X.itemsize)
# print("Tamanho em bytes de cada elemento em Y:", Y.itemsize)
# print("Número total de bytes consumidos por X:", X.nbytes)
# print("Número total de bytes consumidos por Y:", Y.nbytes)
# print("Número de dimensões de X:", X.ndim)
# print("Número de dimensões de Y:", Y.ndim)
```

### Calculando "A" e "B"

```python
# 2. Calculando os coeficientes da regressão linear (a e b)
x_mean = np.mean(X)
y_mean = np.mean(Y)

print("Média de X:", x_mean)
print("Média de Y:", y_mean)
```

### Calculando reta de regressão e Coeficiente de Determinação com numpy()
Nesse bloco de código foram usados métodos do _numpy_ como **polyfit()** e **corrcoef()** para agilizar os cálculos

```python
# Calculando reta de regressão (^y = 𝑎 + 𝑏 ∙ 𝑥)

# Formatando os valores em uma lista(array) para facilitar a conta
x_formatado = dados[:, 0]
y_formatado = dados[:, 1]

# Calcula os coeficientes B (inclinação) e A (intercepto)
# com esses valores poderemos calcular a reta de regressão
b, a = np.polyfit(x_formatado,y_formatado,1)
print(f"Valor de A: {a:.2f}\nValor de B: {b:.2f}")
#[Saída] Valor de A: -22.36
#[Saída] Valor de B: 3.91

# Calculando reta de regressão (^y = 𝑏 ∙ 𝑥 + a)
y_pred = b * x_formatado + a

# Calculando o Coeficiente de Determinação e o seu valor ao quadrado
coef_determinacao = np.corrcoef(x_formatado, y_formatado)[0, 1]
coef_determinacao_2 = coef_determinacao ** 2

# Exibindo resultado
print(f"Coeficiente de Determinação(r2): {coef_determinacao_2 * 100:.2f}%")
#[Saída] Coeficiente de Determinação(r2): 97.51%
```

### Gerando gráfico com matplotlib

```python
# Gerando gráfico com os valores

plt.scatter(X,Y)
plt.plot(x_formatado, y_pred, color="red")

plt.title('Regressão Linear e Coeficiente de Determinação (R²)')
plt.xlabel('Variável independente (X)')
plt.ylabel('Variável Dependente (Y)')

plt.legend(["Dados do CSV", "Linha de Regressão"], loc="lower right")
plt.grid() #Adiciona uma grade ao gráfico
plt.show()
```
