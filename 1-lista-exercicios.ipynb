#### 1) Crie uma classe Aluno que armazene o nome e uma lista de notas. Implemente um método para calcular a média e outro para verificar se o aluno foi aprovado (média >= 6)

```python
class Aluno:
    def __init__(self, input_nome, nota_1, nota_2):
        self.nome = input_nome
        self.notas = [nota_1, nota_2]

    def CacularMedia(self):
        quantidade_notas = len(self.notas)
        soma_notas = 0

        for nota in self.notas:
            soma_notas += nota
        
        return soma_notas / quantidade_notas
    
    def AprovarPorMedia(self, input_media):
        media = float(input_media)

        if(media >= 6):
            return "Aprovado pela média"
        else:
            return "Reprovado pela média"

    
vinicius = Aluno('Vinicius', 9.25, 8.12)
print(f"Média: {vinicius.CacularMedia():.2f}")
print(vinicius.AprovarPorMedia(vinicius.CacularMedia()))
```

#### 2) Crie uma classe Conversor que possua um atributo privado taxa_dolar. Implemente métodos para atualizar a taxa (com validação para não aceitar valores negativos) e para converter um valor de Real para Dólar

```python
class Conversor:
    def __init__(self, input_taxa_dolar):
        self.__taxa_dolar = input_taxa_dolar

    def AtualizarTaxa(self, input_nova_taxa):
        if(input_nova_taxa > 0):
            self.__taxa_dolar = input_nova_taxa

    def ConverterRealParaDolar(self, input_valor_real):
        if(input_valor_real > 0):
            return input_valor_real / self.__taxa_dolar
        
conversor = Conversor(5.25)
conversor.AtualizarTaxa(5.28)
print(f"Convertendo 10 reais para dólar: USD {conversor.ConverterRealParaDolar(10):.2f}")
```

#### 3) Crie as classes Livro e Biblioteca. A classe Biblioteca deve conter uma lista de objetos do tipo Livro. Implemente métodos para adicionar livros e listar todos os títulos disponíveis

```python
class Livro:
    def __init__(self, input_titulo, input_genero, input_autor):
        self.titulo = input_titulo
        self.genero = input_genero
        self.autor = input_autor

    def ExibirInformacoes(self):
        return f"Título: {self.titulo} - Gênero: {self.genero} - Escrito por: {self.autor}"
    
class Biblioteca:
    def __init__(self, input_nome):
        self.nome_biblioteca = input_nome
        self.__livros = []
    
    def AdicionarLivro(self, livro):
        self.__livros.append(livro)
        print(f"{livro.titulo} foi adicionado à coleção da Biblioteca")
    
    def ListarTitulos(self):
        if(len(self.__livros) == 0):
            print("A biblioteca está sem livros no momento")
            return

        print('___ Listando livros ___')
        for livro in self.__livros:
            print(livro.ExibirInformacoes())

biblioteca = Biblioteca('Babilônia')
biblioteca.ListarTitulos()


livro_01 = Livro('A arte da Guerra', 'Estratégia', 'Sun Tzu')
biblioteca.AdicionarLivro(livro_01)

livro_02 = Livro('Roube como um artista', 'Livro de Coach', 'Bruce Lee')
biblioteca.AdicionarLivro(livro_02)

biblioteca.ListarTitulos()
```

#### 4) Crie uma classe base Veiculo com atributos como marca e modelo. Crie subclasses Carro (com número de portas) e Moto (com cilindradas). Ambas devem herdar as características da classe pai

```python
class Veiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def LigarAlarme(self):
        print(f"O alarme do veículo {self.modelo} - {self.marca} foi ligado!")

class Carro(Veiculo):
    def __init__(self, marca, modelo, numero_porta):
        super().__init__(marca, modelo)
        self.numero_porta = numero_porta
    

class Moto(Veiculo):
    def __init__(self, marca, modelo, cilindrada):
        super().__init__(marca, modelo)
        self.cilindrada = cilindrada

fusca = Carro('Volkswagen', 'Fusca 1500', 2)
sahara = Moto('Honda', 'XRE300 Sahara', 300)

fusca.LigarAlarme()
sahara.LigarAlarme()

    
```

#### 7) Crie uma classe Usuario com login e senha. Implemente um método que valide se a senha possui pelo menos 8 caracteres e um caractere especial. Se não atender, o sistema deve solicitar uma nova senha

```python
from string import punctuation

#Caracteres especiais - Créditos: https://docs.python.org/3/library/string.html
#print(punctuation)
#print(set(punctuation))

class Usuario:
    def __init__(self, input_usuario, input_senha):
        self.usuario = input_usuario.strip()
        self.__senha = input_senha.strip()

    def AtualizarSenha(self, input_nova_senha):
        self.__senha = input_nova_senha
        self.ValidarSenha()

    def ValidarSenha(self):
        caracteres_especiais = set(punctuation)
        contem_caractere_especial = False
        minino_caracteres = False

        if(len(self.__senha) < 8):
            print(f'A senha contém {len(self.__senha)} de 8 caracteres mínimos')
        else:
            print('A senha contém a quantidade mínima de caracteres necessários')
            minino_caracteres = True
            

        for letter in self.__senha:
            if letter in caracteres_especiais:
                print('A senha contém pelo menos 1 caractere especial')
                contem_caractere_especial = True
                pass
        
        if(not contem_caractere_especial):
            print('A senha não possui nenhum caractere especial')

        if(not minino_caracteres or not contem_caractere_especial):
            print("Sua senha não foi validada, por favor atualize ela e tente novamente")
        else:
            print("Sua senha tem um formato válido")

usuario = Usuario('vinicius','pass123')  
usuario.ValidarSenha()
print("\n---------------\n")
usuario.AtualizarSenha('vinicius1234')
print("\n---------------\n")
usuario.AtualizarSenha('@1234')
print("\n---------------\n")
usuario.AtualizarSenha('Vinicius@123')
```

#### 8) Crie uma classe Camera e uma classe Telefone. Crie uma terceira classe Smartphone que herde de ambas, demonstrando como o objeto final pode acessar métodos de as duas classes originais.

```python
class Camera:
    def __init__(self):
        pass
    
    def TirarFoto(self):
        print("Foto tirada com sucesso")

    def AjustarLente(self):
        print('A lente da camera foi ajustada')

class Telefone:
    def __init__(self):
        pass

    def FazerLigacao(self, input_contato):
        print(f"Fazendo uma ligação para: {input_contato}")

    def VerHistorico(self):
        print("Visualizando o seu histórico de chamadas")

class Smartphone(Camera, Telefone):
    def __init__(self, input_modelo, input_cor):
        super().__init__()
        self.cor = input_cor
        self.modelo = input_modelo

    def VerDetalhes(self):
        print(f"Detalhes do aparelho - Modelo:{self.modelo} - Cor:{self.cor}")

samsumg = Smartphone('s25', 'Cinza')
samsumg.VerDetalhes()
samsumg.AjustarLente()
samsumg.TirarFoto()
samsumg.VerHistorico()
samsumg.FazerLigacao('Doces du Jorgão')
```

#### 9) Crie uma classe Funcionario que possua um método registrar_entrada(). Utilize o módulo datetime para capturar o horário exato em que o método foi chamado e armazene isso em uma lista de logs dentro do objeto

```python
import datetime

#Referência: https://www.w3schools.com/python/python_datetime.asp

class Funcionario:
    def __init__(self, input_nome):
        self.nome = input_nome
        self.log_entrada = []

    #RegistrarEntrada seria um nome melhor xD
    def registrar_entrada(self):
        hora_ponto = datetime.datetime.now()
        hora_ponto_formatado = hora_ponto.strftime("%H:%M:%S - %d/%m/%Y")
        
        self.log_entrada.append(hora_ponto_formatado)
        print("Entrada registrada: " + hora_ponto_formatado)

    def VisualizarLog(self):
        print(f'___ Visualizando logs do funcionário {self.nome} ___')

        if(len(self.log_entrada) > 0):
            for log in self.log_entrada:
                print(log)
        else:
            print("Sem registros no momento")

funcionario = Funcionario('Marcos')

funcionario.registrar_entrada()
funcionario.registrar_entrada()
funcionario.registrar_entrada()
funcionario.registrar_entrada()
funcionario.registrar_entrada()

funcionario.VisualizarLog()
```

#### 10) Crie uma classe Calculadora que contenha apenas métodos estáticos para operações como soma, subtração e raiz quadrada. O objetivo é usar a classe como um "namespace" de funções utilitárias sem precisar instanciá-la.

```python
class Calculadora:
    def __init__(self):
        pass

    def Soma(num1, num2):
        return num1 + num2
    
    def Subtracao(num1, num2):
        return num1 - num2
    
    def RaizQuadrada(num1):
        return num1**2

print(f"A operação 2 + 5 é igual a {Calculadora.Soma(2, 5)}")
print(f"A operação 4 - 6 é igual a {Calculadora.Subtracao(4, 6)}")
print(f"A operação 9 ao quadrado é igual a {Calculadora.RaizQuadrada(9)}")
```
